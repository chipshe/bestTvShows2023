```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# URL
URL = 'https://www.imdb.com/list/ls562785392/'

# Sayfa Bağlantısı
page = requests.get(URL)
content = BeautifulSoup(page.content, 'html.parser')
```

```python
# Serie Index, Name, Year
serieName = content.find_all('h3', class_ = 'lister-item-header')

serieNameList = []

for item in serieName:
    # Serie Index
    serieIndex = item.find('span', class_='lister-item-index').get_text(strip=True)
    
    # Serie Title
    serieTitle = item.find('a').get_text(strip=True)
    
    # Serie Year
    serieYear = item.find('span', class_='lister-item-year').get_text(strip=True)
    
    # Append to the list
    serieNameList.append({'Rank': serieIndex, 'Title': serieTitle, 'Year': serieYear})

# Print the result
for serie in serieNameList:
    print(f"{serie['Rank']}. {serie['Title']} ({serie['Year']})")
```

```python
# DataFrame
df = pd.DataFrame(serieNameList)
df.set_index('Rank', inplace = True)
df.head(10)
```

```python
# Serie Index, Name, Year, Rating
serieName = content.find_all('h3', class_='lister-item-header')
serieRatings = content.find_all('div', class_='ipl-rating-widget')

serieNameList = []

for idx, item in enumerate(serieName):
    # Serie Index
    serieIndex = item.find('span', class_='lister-item-index').get_text(strip=True)
    
    # Serie Title
    serieTitle = item.find('a').get_text(strip=True)
    
    # Serie Year
    serieYear = item.find('span', class_='lister-item-year').get_text(strip=True)

    # Serie Rating
    serieRating = serieRatings[idx].find('span', class_='ipl-rating-star__rating').get_text(strip=True)

    # Append to the list
    serieNameList.append({'Rank': serieIndex, 'Title': serieTitle, 'Year': serieYear, 'IMDB Rating': serieRating})

# Print the result
for serie in serieNameList:
    print(f"{serie['Rank']}. {serie['Title']} ({serie['Year']}) ({serie['IMDB Rating']})")
```

```python
# DataFrame
df = pd.DataFrame(serieNameList)
df.set_index('Rank', inplace = True)
df.head(10)
```

```python
# Serie Index, Name, Year, Rating
serieName = content.find_all('h3', class_='lister-item-header')
serieRatings = content.find_all('div', class_='ipl-rating-widget')
plotSummary = content.find_all('p', class_ = '')

serieNameList = []

for idx, item in enumerate(serieName):
    # Serie Index
    serieIndex = item.find('span', class_='lister-item-index').get_text(strip=True)
    
    # Serie Title
    serieTitle = item.find('a').get_text(strip=True)
    
    # Serie Year
    serieYear = item.find('span', class_='lister-item-year').get_text(strip=True)

    # Serie Rating
    serieRating = serieRatings[idx].find('span', class_='ipl-rating-star__rating').get_text(strip=True)

    # Plot Summary
    pSum = plotSummary[idx].get_text(strip = True)

    # Append to the list
    serieNameList.append({'Rank': serieIndex, 'Title': serieTitle, 'Year': serieYear, 'IMDB Rating': serieRating, 'Plot Summary': pSum})

# Print the result
for serie in serieNameList:
    print(f"{serie['Rank']}. {serie['Title']} ({serie['Year']}) ({serie['IMDB Rating']}) ({serie['Plot Summary']})")
```

```python
# DataFrame
df = pd.DataFrame(serieNameList)
df.set_index('Rank', inplace = True)
df
```

```python
# CSV 
df.to_csv('serie_data.csv', encoding='utf-16', index=True)
```
